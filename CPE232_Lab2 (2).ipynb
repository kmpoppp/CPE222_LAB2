#Time-Series Data#

```python
df = pd.read_csv('/content/drive/MyDrive/Electric_Production.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
date_column = 'DATE'
value_column = 'IPG2211A2N'

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')

sns.lineplot(x=date_column, y=value_column, data=df)

plt.xlabel(date_column)
plt.ylabel(value_column)
plt.title('Time Series Plot of ' + value_column)

sns.despine(fig=fig, ax=ax)
plt.show()
```

#Semi-Structured#

```python
import requests
import pandas as pd
```

```python
url = "https://min-api.cryptocompare.com/data/histoday?fsym=BTC&tsym=ETH&limit=30&aggregate=1&e=CCCAGG" ###nested table
resp = requests.get(url)
```

```python
data = resp.json()['Data']
```

```python
TabWOData = resp.json()
del TabWOData['Data']
TabWOData
```

```python
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df2 = pd.DataFrame(TabWOData)
df2
```

```python
df2.info()
```

```python
df2.describe()
```

```python
df3 = pd.concat([df,df2], axis =1)
df3
```
